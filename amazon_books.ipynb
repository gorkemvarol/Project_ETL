```python
#dependencies
from bs4 import BeautifulSoup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import datetime
```

```python
current_year = datetime.datetime.now()
date = current_year.date()
year = int(date.strftime("%Y"))
```

```python
scraped = pd.DataFrame()
```

```python
#browser
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#specify page url
url="https://www.amazon.com/gp/bestsellers/1994/books/ref=zg_bs_tab_t_bsar_pg_1"
browser.visit(url)
print(url)
```

```python
#save html per page to varable and parse
record = 1
for x in range(1994,year+1):
    for page in range(2):
        print(url)
        html = browser.html
        soup = BeautifulSoup(html, 'html.parser')

        books = soup.find_all('li', class_='zg-item-immersion')

        for book in books:
            try:
                a = book.find('a')
                href = a['href']
                link = 'https://www.amazon.com'+href
                browser.visit(link)
                book_html = browser.html
                book_soup = BeautifulSoup(book_html, 'html.parser')
                info_dict = {}
                book_title = book_soup.find('span',class_='a-size-extra-large').text[1:]
                try:
                    book_author = book_soup.find('a',class_='a-link-normal contributorNameID').text
                except:
                    AttributeError
                    try:
                        book_author = book_soup.find('a',class_='a-link-normal').text
                    except:
                        AttributeError
                book_info = book_soup.find_all('li', class_='a-carousel-card rpi-carousel-attribute-card')
                info_dict.update({'title':book_title, 'author':book_author})
                for element in book_info:
                    span = element.find_all('span')
                    param = span[0].text
                    info = span[2].text
                    info_dict.update({param:info})
                scraped = scraped.append(info_dict, ignore_index=True) 
                print(record)
                record+=1
            except:
                AttributeError
                print(record)
                
        browser.visit(url)
        try:
            print(f"---{x}---")
            browser.links.find_by_partial_text('Next page').click()
            url = f'https://www.amazon.com/gp/bestsellers/{x}/books/ref=zg_bsar_pg_2?ie=UTF8&pg=2'
            
        except:
            print(f"---{x}---")
            url=f'https://www.amazon.com/gp/bestsellers/{x}/books/ref=zg_bs_tab_t_bsar_pg_1'
            browser.visit(url)

```
